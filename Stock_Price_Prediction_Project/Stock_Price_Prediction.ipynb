```python
import FinanceDataReader as fdr
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 삼성전자(005930), 카카오 (035720), 하이브(352820), 더네이쳐홀딩스(298540), 코엔텍(029960)
START_DATE = '2022-11-01'
END_DATE = '2023-04-21'
TICKER_LIST = ['005930', '035720', '352820', '298540', '029960']
COLUMNS_NAME = ['삼성전자', '카카오', '하이브', '더네이쳐홀딩스', '코엔텍']
PREIDCTION_DATE = [20230424,20230425,20230426,20230427,20230428]

def train_dataset_loading(ticker_list, columns_name, start_date, end_date):
    dt = pd.DataFrame()
    for i in ticker_list:
        stock = fdr.DataReader(i, start_date, end_date)
        stock_close = stock['Close']
        dt = pd.concat([dt, stock_close], axis=1)
    dt.columns = columns_name
    return dt


train_dataset = train_dataset_loading(TICKER_LIST,COLUMNS_NAME, START_DATE, END_DATE)
train_dataset.head()
```

## AutoArima

```python
from pmdarima.arima import auto_arima

def autoarima(dataframe):
    model = auto_arima(dataframe,seasonal=True, trace=True, error_action='ignore', suppress_warnings=True)
    model.fit(dataframe)
    forecast = model.predict(n_periods=5)
    return forecast


def predict_to_dtframe(train_dataset, model, seperate=True):
    
    predict = pd.DataFrame()
    if seperate: 
        for i in range(len(train_dataset.columns)):
            forecast = model(train_dataset.iloc[:,i])
            predict[train_dataset.columns[i]] = forecast
            
    else:
        forecast = model(train_dataset)
        predict = pd.DataFrame(forecast)
        predict.columns = train_dataset.columns
        
    predict['Date'] = PREIDCTION_DATE
    predict = predict[['Date',*COLUMNS_NAME]]
    predict.index = list(range(0,5))
    return predict

predict_ARIMA = predict_to_dtframe(train_dataset, autoarima)

```

## Test_Data_Loading

```python
def test_dataset_loading ():
    stocks = []
    for i in ['005930', '035720', '352820', '298540', '029960'] : 
        stock = fdr.DataReader(i, '2023-04-24', '2023-04-28')
        for j in range(5) :
            stocks.append(stock['Close'][j])
    submission = pd.DataFrame()
    submission['Date'] = PREIDCTION_DATE
    submission['삼성전자'] = stocks[:5]
    submission['카카오'] = stocks[5:10]
    submission['하이브'] = stocks[10:15]
    submission['더네이쳐홀딩스'] = stocks[15:20]
    submission['코엔텍'] = stocks[20:25]
    return submission
submission = test_dataset_loading()
```

```python
# 오른쪽 항은 예측값을 넣으시면 됩니다!
## 아래의 오른쪽 항은 산식 산출 확인을 위해 임의로 넣은 값입니다!

def submission_prediction(submission, predict, model_name='ARIMA'):
    submission[f'삼성전자_{model_name}'] = predict['삼성전자']
    submission[f'카카오_{model_name}'] = predict['카카오']
    submission[f'하이브_{model_name}'] = predict['하이브']
    submission[f'더네이쳐홀딩스_{model_name}'] = predict['더네이쳐홀딩스']
    submission[f'코엔텍_{model_name}'] = predict['코엔텍']
    return submission

## 이 아래는 수정 사항 없습니다.

def final_score(submission, model_name='ARIMA'):
    a = np.sum(np.abs(submission['삼성전자']-submission[f'삼성전자_{model_name}']))/np.sum(submission['삼성전자'])
    b = np.sum(np.abs(submission['카카오']-submission[f'카카오_{model_name}']))/np.sum(submission['카카오'])
    c = np.sum(np.abs(submission['하이브']-submission[f'하이브_{model_name}']))/np.sum(submission['하이브'])
    d = np.sum(np.abs(submission['더네이쳐홀딩스']-submission[f'더네이쳐홀딩스_{model_name}']))/np.sum(submission['더네이쳐홀딩스'])
    e = np.sum(np.abs(submission['코엔텍']-submission[f'코엔텍_{model_name}']))/np.sum(submission['코엔텍'])
       

    final_score = np.mean([a,b,c,d,e]) * 100
    return final_score


submission = submission_prediction(submission, predict_ARIMA)
final_score(submission)
```

```python
test_dataset = submission.iloc[:,:6]
test_dataset.Date = test_dataset.Date.astype('str').astype('datetime64[D]')
test_dataset = test_dataset.set_index('Date')

test_dataset
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 데이터 로드

# 데이터 전처리
scaler = MinMaxScaler()  # 데이터 스케일링을 위한 scaler 객체 생성
scaled_data_train = scaler.fit_transform(train_dataset)  # 데이터 스케일링
scaled_data_test = scaler.transform(test_dataset)

scaled_data = np.append(scaled_data_train, scaled_data_test, axis=0)



# 학습 데이터와 타깃 데이터 분리
lookback = 20  # 20개의 시점을 사용하여 다음 시점을 예측
X = []
y = []
for i in range(len(scaled_data) - lookback):
    X.append(scaled_data[i:i+lookback])
    y.append(scaled_data[i+lookback])
X = np.array(X)
y = np.array(y)

# 학습 데이터와 테스트 데이터 분리
split = -5
X_train = X[:split]
y_train = y[:split]
X_test = X[split:]
y_test = y[split:]

# LSTM 모델 구성
model = Sequential()
model.add(LSTM(64, input_shape=(lookback, 5)))  # 입력 형태는 (시점 수, 주식 종목 수)
model.add(Dropout(0.2))  # Dropout 추가
model.add(Dense(5))  # 출력은 주식 종목 수와 동일


# 모델 컴파일 및 학습
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=200, batch_size=32)

# 테스트 데이터로 예측 수행
predictions = model.predict(X_test)

# 예측 결과 복원
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform(y_test)

# 예측 결과 출력
for i in range(len(predictions)):
    print('실제값:', y_test[i])
    print('예측값:', predictions[i])
    print("MSE:", np.mean(np.square(y_test[i] - predictions[i])))
    print('---')
```

```python
predict_LSTM = pd.DataFrame(predictions, columns=COLUMNS_NAME, index=list(range(5)))
submission = submission_prediction(submission, predict_LSTM, model_name='LSTM')
final_score(submission, model_name='LSTM')
```
